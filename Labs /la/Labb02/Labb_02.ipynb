```python
import matplotlib.pyplot as plt
import math

# Read training data
data = []
with open('datapoints.txt') as file:
    next(file)
    for line in file:
        width, height, label = line.strip().split(',')
        data.append({
            'width': float(width), 
            'height': float(height), 
            'label': int(label),
        })

# User input and error handling
def get_user_point():
    while True:
        try:
            user_input = input("Enter a test point as 'width, height' (e.g., '25, 32'): ")
            width_str, height_str = user_input.split(',')
            
            width = float(width_str.strip())
            height = float(height_str.strip())
            
            if width < 0 or height < 0:
                print("Error: Please enter positive numbers only.")
                continue
                
            return {'width': width, 'height': height}
            
        except ValueError:
            print("Error: Please enter valid numbers in the format 'width, height'.")
        except Exception as e:
            print(f"Error: Invalid input format. Please use 'width, height' format.")

# Get user input
user_test_point = get_user_point()

# k-NN k=10
distances_with_labels = []
for data_p in data:             
    d = math.dist([user_test_point['width'], user_test_point['height']], 
                  [data_p['width'], data_p['height']]) 
    distances_with_labels.append((d, data_p['label']))

# Sort and get 10 closest points
distances_with_labels.sort(key=lambda x: x[0])
k_nearest = distances_with_labels[:10]

# Count votes
pichu_votes = sum(1 for dist, label in k_nearest if label == 0)
pikachu_votes = len(k_nearest) - pichu_votes

# Majority vote
if pichu_votes > pikachu_votes:
    user_test_point['label'] = 0
    user_test_point['class'] = 'Pichu'
    print(f"Classification: Pichu (votes: {pichu_votes}-{pikachu_votes})")
elif pikachu_votes > pichu_votes:
    user_test_point['label'] = 1
    user_test_point['class'] = 'Pikachu'
    print(f"Classification: Pikachu (votes: {pichu_votes}-{pikachu_votes})")
else:
    user_test_point['label'] = 1  # Tie: Pikachu
    user_test_point['class'] = 'Pikachu'
    print(f"Classification: Pikachu - tie breaker (votes: {pichu_votes}-{pikachu_votes})")

# User point to test
test_points = [user_test_point]

# Plotting
pichu_points = [points for points in data if points['label'] == 0]
pikachu_points = [points for points in data if points['label'] == 1]

pichu_widths = [points['width'] for points in pichu_points]
pichu_heights = [points['height'] for points in pichu_points]
pikachu_widths = [points['width'] for points in pikachu_points]
pikachu_heights = [points['height'] for points in pikachu_points]

plt.scatter(pichu_widths, pichu_heights, color='gold', label='Pichu (training)')
plt.scatter(pikachu_widths, pikachu_heights, color='firebrick', label='Pikachu (training)')

# Plot user point 
plt.scatter([user_test_point['width']], [user_test_point['height']], 
           color='cyan', marker='o', s=50, label=f'User point: {user_test_point["class"]}')

plt.title('Pichu vs Pikachu')
plt.xlabel('Width (cm)')
plt.ylabel('Height (cm)')
plt.legend()
plt.show()
```
