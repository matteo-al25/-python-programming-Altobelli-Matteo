```python
import matplotlib.pyplot as plt
import re

## https://www.geeksforgeeks.org/python/how-to-create-a-python-dictionary-from-text-file/#using-dict
## reading in the data and create a dictionary
data = []
with open('datapoints.txt') as file:
    next(file)
    for line in file:
        width, height, label = line.strip().split(',')
        data.append({
            'width': float(width), 
            'height': float(height), 
            'label': int(label),
        })


## Separate Pichu and Pikachu 
pichu_points = [points for points in data if points['label'] == 0]
pikachu_points = [points for points in data if points['label'] == 1]

## https://www.geeksforgeeks.org/python/matplotlib-pyplot-scatter-in-python/
## Extract coordinates 
pichu_widths = [points['width'] for points in pichu_points]
pichu_heights = [points['height'] for points in pichu_points]
pikachu_widths = [points['width'] for points in pikachu_points]
pikachu_heights = [points['height'] for points in pikachu_points]

# Create scatter plot
plt.scatter(pichu_widths, pichu_heights, color='yellow', label='Pichu')
plt.scatter(pikachu_widths, pikachu_heights, color='red', label='Pikachu')

# Add text and legend
plt.title('Pichu vs Pikachu scatterplot')
plt.xlabel('Width (cm)')
plt.ylabel('Height (cm)')
plt.legend()
plt.show()

# Read and filter the testpoints
filtered = []
file = open('testpoints.txt', 'r')
read = file.readlines()
file.close()  # Close the file after reading

# Remove spaces and 1st line
for line in read:
    if line.startswith('T'):
        continue
    filtered.append(line.strip())

# Remove ordinal numbers
ordinal = []
for point in filtered:
    cleaned = point.split('. ', 1)[-1]
    ordinal.append(cleaned)

# Remove parentheses
parentheses = []
for point in ordinal:
    no_parentheses = point.replace('(', '').replace(')', '')
    parentheses.append(no_parentheses)

# Split values
split_txt = []
for point in parentheses:
    values = point.split(',')
    split_txt.append(values)

# Convert to float
test_points = []
for point in split_txt:
    width = float(point[0].strip())
    height = float(point[1].strip())
    test_points.append({'width': width, 'height': height})

print(test_points)
```
