## Area **[6.1]**

Create a function that takes the base and height of a triangle as input parameters and returns the area of the triangle.

````python
def area():
    base = int(input("Enter the base of the triangle: "))
    height = int(input("Enter the height of the triangle: "))
    return (base * height) / 2

triangle = area()
print(f"The area of the triangle is: {triangle}")
````

## Euclidean distance **[6.2]**

The formula for Euclidean distance in 2D between $P: (p_1, p_2)$ and $Q: (q_1, q_2)$  is: 

$d(P,Q) = \sqrt{(p_1-q_1)^2 + (p_2-q_2)^2}$

&nbsp; a) &nbsp; Create a function that takes two points as input parameters and return the Euclidean between them. (*)

&nbsp; b) &nbsp; Let the user input two points. Call the function using the users input points. (*)

&nbsp; c) &nbsp; Use your function to calculate distances between the origin (0, 0) and each of these points: (10, 3), (-1, -9), (10, -10), (4, -2), (9, -10).

````python
import math

# returning Euclidean
def euclidean(p1, p2, q1, q2):
    dist = math.sqrt((p1 - q1)**2 +(p2 - q2)**2)
    return dist

# using user input
def distance():
    p1 = float(input("Enter p1: "))
    p2 = float(input("Enter p2: "))
    q1 = float(input("Enter q1: "))
    q2 = float(input("Enter q2: "))
    dist = math.sqrt((p1 - q1)**2 +(p2 - q2)**2)
    return dist

result = distance()
print(f"The Euclidean distance between your points is: {result:.2f}")

# calculate distances 
points = [(10, 3), (-1, -9), (10, -10), (4, -2), (9, -10)]

z = []
for x, y in points:
    result = euclidean(0, 0, x, y)
    z.append(result)

print("Distances from the origin 0.0:")
for distance in z:
    print(f"  {distance:.2f}")
````

## Mathematical functions **[6.3]**

Make the following functions with **def** or **lambda** and plot their graphs in the same figure window, with $x\in [-10,10]$ :

&nbsp; a) &nbsp;  $f(x) = x^2 -3$ (*)

&nbsp; b) &nbsp; $g(x) = 4x-7$ (*)

What could the relation between $f(x)$ and $g(x)$ be?


````python
import matplotlib.pyplot as plt
import numpy as np

# lambda functions
f = lambda x: x**2 - 3  
g = lambda x: x*4 -7
x = np.linspace(-10, 10)

# finding y
y_f = f(x)
y_g = g(x)

# creating line plot
plt.plot(x, y_f, label="f(x) = x² - 3")
plt.plot(x, y_g, label="g(x) = 4x - 7")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Functions f(x) and g(x)")
plt.legend()
plt.grid(True)
plt.show()

# finding relation between the 2 functions
relation = x**2 -x*4 + 4
intersection_x = x[np.argmin(np.abs(relation))]
intersection_y = f(intersection_x)
print(f"The functions intersect at ({intersection_x:.2f}, {intersection_y:.2f})")
````

## Name cleaner **[6.4]**

Create a function that takes a name as an input and:

- removes all leading and trailing blank spaces
- make capitalize the first character of each name, and make the rest lowercase

Use your function on this list of strings: 

```
["  MaRcUs ", " iDA aNderSon", "OLOF Olofsson            "  ]

````python
names = ["  MaRcUs ", " iDA aNderSon", "OLOF Olofsson            "]
for name in names:
    words = name.strip().split()
    capitalized_words = [word.title() for word in words]
    cleaned_name = ' '.join(capitalized_words)
    print(cleaned_name)
````
